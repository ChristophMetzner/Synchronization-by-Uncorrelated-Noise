```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import os
import pickle
```

```python
with open('.test_cond.pkl', 'rb') as handle:
    data = pickle.load(handle)

print(data['model_results']['net'].keys())
```

```python
# rates
r_e1 = data['model_results']['net']['r_e']
r_e2 = data['model_results']['net']['r_e2']
r_i1 = data['model_results']['net']['r_i1']
r_i2 = data['model_results']['net']['r_i2']

# spikes
s_e1 = data['model_results']['net']['net_spikes_e']
s_e2 = data['model_results']['net']['net_spikes_e2']
s_i1 = data['model_results']['net']['net_spikes_i1']
s_i2 = data['model_results']['net']['net_spikes_i2']

# time
t = np.linspace(0,3000,int(3000/0.05))
```

```python
fig = plt.figure(figsize=[20,15])
ax  = fig.add_subplot(111)
ax.plot(t,r_e1,'0.75')
ax.plot(t,r_e2,'0.25')


fig = plt.figure(figsize=[20,15])
ax  = fig.add_subplot(111)
ax.plot(t,r_i1,'0.75')
ax.plot(t,r_i2,'0.25')
```

```python
fig = plt.figure(figsize=[20,15])
ax  = fig.add_subplot(111)
ax.plot(s_e1[1]*1000,s_e1[0],'k.')
```
