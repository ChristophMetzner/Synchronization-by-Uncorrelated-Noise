```python
import analysis
import plots
import runner
from constants import PLOTS_PATH

%load_ext autoreload
%autoreload 2
```

# Exploration of variations of noise input 

```python
params = dict()
params['runtime'] = 400
params['min_dt'] = 0.05
params['ou_stationary'] = True
```

### Weak Correlation - Default Configuration

```python
params['ou_mu'] = {
    'ou_X0': 0.,
    'ou_mean': 5.0,
    'ou_sigma': .5,
    'ou_tau': 50.
}

# noise params for sigma
params['ou_sigma'] = {
    'ou_X0': 0.,
    'ou_mean': 2.0,
    'ou_sigma': 0.2,
    'ou_tau': 1.
}

fig, ax = plots.ou_noise_by_params(params)
fig.savefig(f"{PLOTS_PATH}/noise_experiment/weak_noise_correlation")
```

### Medium Correlation - Default Configuration

```python
params['ou_mu'] = {
    'ou_X0': 0.,
    'ou_mean': 5.0,
    'ou_sigma': .5,
    'ou_tau': 500.
}

# noise params for sigma
params['ou_sigma'] = {
    'ou_X0': 0.,
    'ou_mean': 2.0,
    'ou_sigma': 0.2,
    'ou_tau': 1.
}

fig, ax = plots.ou_noise_by_params(params)
fig.savefig(f"{PLOTS_PATH}/noise_experiment/medium_noise_correlation")
```

### Very Strong Correlation over Time

```python
params['ou_mu'] = {
    'ou_X0': 0.,
    'ou_mean': 5.0,
    'ou_sigma': .5,
    'ou_tau': 5000.
}

params['ou_sigma'] = {
    'ou_X0': 0.,
    'ou_mean': 2.0,
    'ou_sigma': 0.2,
    'ou_tau': 10000.
}

fig, ax = plots.ou_noise_by_params(params)
fig.savefig(f"{PLOTS_PATH}/noise_experiment/strong_noise_correlation")
```

```python
results = runner.run("strong_correlation", modified_params=params)
```

```python
analysis.analyze_model(results, 'strong_correlation')
```
